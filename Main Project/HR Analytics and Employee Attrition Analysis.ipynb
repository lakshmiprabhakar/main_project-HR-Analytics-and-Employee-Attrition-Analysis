<h4 style="color:#2E8B57;">
8.6 Create New Columns
</h4>
In this step, new meaningful features were created from existing data to improve analysis and extract deeper insights from the dataset.

```python
##1. Creating Age Group Column
```

```python
def age_group(age):
    if age < 30:
        return 'Young'
    elif age < 40:
        return 'Adult'
    else:
        return 'Senior'

df['agegroup'] = df['age'].apply(age_group)
```

#### Explanation

apply()  
- Applies custom function to every value  

Why Used?
- To categorize employees into age groups  

New Column Created:
- agegroup

```python
##2. Creating Income Level Column
```

```python
def income_level(income):
    if income < 5000:
        return 'Low Income'
    elif income < 10000:
        return 'Medium Income'
    else:
        return 'High Income'

df['incomelevel'] = df['monthlyincome'].apply(income_level)
```

#### Explanation

Employees are categorized based on salary level for income analysis.

```python
##3. Creating Experience Level Column
```

```python
def experience_level(exp):
    if exp < 5:
        return 'Fresher'
    elif exp < 10:
        return 'Experienced'
    else:
        return 'Highly Experienced'

df['experienceLevel'] = df['totalworkingyears'].apply(experience_level)
```

#### Explanation
Employees are grouped based on work experience.

```python
##4. Creating Satisfaction Level Column
```

```python
def satisfaction_level(x):
    if x <= 2:
        return 'Low Satisfaction'
    else:
        return 'High Satisfaction'

df['satisfactionlevel'] = df['jobsatisfaction'].apply(satisfaction_level)
```

#### Explanation

Employees are classified based on satisfaction score.

```python
## Dataset After Feature Engineering
```

```python
df.head()
```

1. Age Group — Categorizes employees into life-stage buckets (Young, Adult, Senior) based on their age, making it easier to analyze attrition patterns across generational cohorts rather than individual ages.

2. Income Level — Groups employees by salary band (Low Income, Medium Income, High Income) derived from their DailyRate or monthly equivalent, simplifying compensation-based analysis.

3. Experience Level — Segments employees by tenure or professional maturity (Entry-Level, Experienced, Highly Experienced), helping identify whether time in the workforce correlates with turnover.

4. Satisfaction Level — Aggregates JobSatisfaction and EnvironmentSatisfaction into ordinal categories (Low, Medium, High Satisfaction), providing a single usable metric for tracking how employee sentiment relates to attrition.

#### Overtime & Attrition — Analysis

**The Big Picture**

Out of 1,470 total employees in the dataset, 416 worked overtime and 1,054 did not. Overall, 237 employees left the company — an attrition rate of 16.1%.

**The Overtime Difference**

Among employees who did not work overtime, only 110 out of 1,054 left — that's an attrition rate of 10.4%, or roughly 1 in 10.

Among employees who did work overtime, 127 out of 416 left — that's an attrition rate of 30.5%, or roughly 3 in 10.

**What This Means**

An overtime worker is about three times more likely to leave compared to a non-overtime worker. Furthermore, although overtime workers make up only 28% of the workforce, they account for more than half (54%) of all employees who left.

**Bottom Line**

There is a strong link between working overtime and attrition. The data clearly suggests overtime is a risk factor — possibly driven by burnout, poor work-life balance, or dissatisfaction with overtime expectations. This warrants deeper investigation into why overtime employees are leaving at such a higher rate and whether reducing mandatory overtime could improve retention.
